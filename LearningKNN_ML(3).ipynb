```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = dt.iloc[:, [1,2]].values
y = dt.iloc[:, 3].values
```

variabel x = memuat atribute
variabel y = memuat label/target
[] =mengambil semua baris
[1,2] dan [-1] = mengambil kolom atau index tertentu

```python
from sklearn.model_selection import train_test_split # Corrected import statement
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 0)
```

```python
print(x_train)
```

```python
print(x_test)
```

Scalling Data
bedanaya min max sm standar s-score adalah min max dipakai saat jumlah nilai dalam data itu rentang ny tidak jauh banyak, kalau s-score itu untuk nilai data yg rentang ny jauh

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) # Corrected argument name to n_neighbors
knn.fit(x_train, y_train)

y_predict = knn.predict(x_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
print(cm)
```

```python
# Cek Akurasi
from sklearn.metrics import classification_report
akurasi = classification_report(y_test, y_predict)
print(akurasi)

from sklearn.metrics import accuracy_score
akurasi = accuracy_score(y_test, y_predict)
print('Tingkat Akurasi Adalah: %d persen'%(akurasi*100))
```

```python
from matplotlib.colors import ListedColormap
x_set, y_set = x_train, y_train

x1, x2 = np.meshgrid(
    np.arange(start=x_set[:,0].min() - 1, stop=x_set[1,0].max() + 1, step=0.01),
    np.arange(start=x_set[:,1].min() - 1, stop=x_set[1,1].max() + 1, step=0.01)
    )

plt.contourf(
    x1, x2, knn.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
    alpha=0.75, cmap=ListedColormap(('red', 'blue'))
)

for i, j in enumerate(np.unique(y_set)):
  plt.scatter(
      x_set[y_set == j, 0], x_set[y_set == j, 1],
      color=ListedColormap(('red', 'blue'))(i), label=j
  )

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
plt.xlabel('Fitur 1 (Setelah Standarisasi)')
plt.ylabel('Fitur 2 (Setelah Standarisasi)')
plt.legend()
plt.title('Decision Boundary dengan KNN (Sandarization)')
plt.show()
```
